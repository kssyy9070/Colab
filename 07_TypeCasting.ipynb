# 1-1. 단위 환산 : round()

```python
t = sns.load_dataset('titanic')
t.head()
```

```python
# rount() : 반올림 함수
t['fare'] = t['fare'].round(2) # 괄호안에 소수점 자릿수
t.head()
```

```python
# round()함수는 데이터프레임.round(소수점자리수) / round(데이터프레임.소수점자리수) 두가지 방법 모두 동일
# 괄호안에 아무 숫자도 넣지 않으면, 소수점 첫번째자리에서 반올림
print(round(1.6667))
print(round(1.6667,3))
```

```python
a = pd.read_excel('myStudent.xlsx')
a.head()
```

```python
# 국어, 영어, 수학 점수 다 더한값의 평균값 계산
a['avg'] = (a['kor']+a['eng']+a['math']) / 3
a.head()
```

```python
a.dtypes
```

```python
a['gender'] = a['gender'].astype('category')
a.dtypes
```

```python
# 카테고리로 바꾸기
a['group'] = a['group'].astype('category')
a['education'] = a['education'].astype('category')
a.dtypes
```

```python
a[['gender','group','education']].describe()
```

```python
# gender가 f,m으로만 나타나는데 female로 바꾸려고 함

# replace() 함수 : 값 변경
# 다양한 값을 한번에
# replace( {변경전 : 변경후, ...} )
a['gender'].replace({'f':'female','m':'male'},inplace=True)
a['gender'].unique()
```

# 2. 범주형(카테고리) 데이터 처리

# 2-1. 구간 분할 : np.histogram(), pd.cut()

```python
aa = pd.DataFrame(a)
aa.head()
```

```python
# np.histogram() 함수 : np.histogram(나눌 리스트, bins = 나눌 구간)의 형태
# 2개의 값 :
#   count : 각 구간에 속하는 값의 개수
#   bins_diveders : 경계값 리스트를 변환
count, bins_dividers = np.histogram(aa['avg'],bins=3)
print(count,'\n',bins_dividers)
```

```python
# pd.cut()함수 : 자동으로 범주형변수를 생성해주고 간단함
bin_names = ['low','middle','high']
aa['aa_bin'] = pd.cut(x=aa['avg'],          # 데이터 배열
                      bins=bins_dividers,   # 경계값 리스트
                      labels = bin_names,   # 구간명
                      include_lowest = True) # 첫 경계값 포함여부(구간의 하위값)

aa.head()
```
