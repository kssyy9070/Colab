```python
# 필요한 라이브러리 불러오기
from bs4 import BeautifulSoup
import urllib.request as req
import requests
import pandas as pd
```

```python
# https://www.menupan.com/restaurant/bestrest/bestrest.asp?pt=rt&areacode=sn217

# 전체 내용 파싱
# WebCrawing

# 1
url = "https://www.menupan.com/restaurant/bestrest/bestrest.asp?pt=rt&areacode=sn217"
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html5lib')
print(soup)

# 2
# url = "https://www.menupan.com/restaurant/bestrest/bestrest.asp?pt=rt&areacode=sn217"
# response = req.urlopen(url)
# # soup = BeautifulSoup(response,'html.parser')  #html.parser, html5lib, lxml
# soup = BeautifulSoup(response,'html5lib')       # 소스를 보기 편하게 정렬 그러나 처리속도가 약간 느림
# print(soup)
```

```python
# 순위, 상호, 업종, 주소, url에 대한 내용 출력
# select() vd select_one()

# select() : 매칭되는 여러 값들을 리스트에 담음
# select_one() : 매칭되는 값 하나만 가져와서 => 객체

# select 관련 => 하나의 열에 여러 선택자가 필요한 경우
#     => .000 or .XXX => (or : ','(쉼표))

# 속성값 구해야 할 때 : [] 사용

# body > div > div.container > div.bestRest > div.ranking > div.rankingList > ul

s = 'body > div > div.container > div.bestRest > div.ranking > div.rankingList > ul'
ulTag = soup.select_one(s)
# print(ulTag)

for li in ulTag.select('li'):
  print('순위 : ',li.select_one('.numTop, .rankNum').string)
  print('상호 : ',li.select_one('.restName').string)
  print('업종 : ',li.select_one('.listType').string)
  print('주소 : ',li.select_one('.listArea').string)
  # print('url : ',li.select_one('.listName a')['href']) # 이름 클릭했을때 이동하게 되는
  print('url : ',li.select_one('.listPhoto a')['href'])  # 사진 클릭했을때 이동하게 되는
  # img의 경로지정 속성은 src(그 경로에 있는 이미지를 불러오겠다.)
  # a 태그의 url 경로 지정 속성은 href (그 경로로 이동을 하겠다.)
  print('----------------------------')
```

```python
# 위의 데이터들을 list로 나타내는 작업 

menuList = [] # 각 객체를 dict 형태로 // [{'순위' : '1', '상호' : 'xxx', '업종' : '한식', ....},....]
urlList = []  # url 

s = 'body > div > div.container > div.bestRest > div.ranking > div.rankingList > ul'
ulTag = soup.select_one(s)

for li in ulTag.select('li'):
  menuList.append({'순위' : li.select_one('.numTop, .rankNum').string,
                   '상호' : li.select_one('.restName').string,
                   '업종' : li.select_one('.listType').string,
                   '주소' : li.select_one('.listArea').string
  })
  urlList.append(li.select_one('.listPhoto a')['href'])

# menuList
# urlList
```

```python
# menuList를 가지고 순위를 index로 하는 DataFrame을 생성


menuDF = pd.DataFrame(menuList)
menuDF.set_index('순위',inplace=True)
menuDF
```

```python
# https://www.menupan.com/ 이 기본 주소값 뒤에 아까 담아놓은 주소를 붙여서
# 전화번호, 상세주소 를 가져온 후에 
# 만들어 놓은 DataFrame에 붙이기 ('주소' 열은 삭제)

baseUrl = 'https://www.menupan.com'
telList = []
addrList = []

for u in urlList:
  response = requests.get(baseUrl+u)
  soup = BeautifulSoup(response.content, 'html5lib')
  telList.append(soup.select_one('.tel1').string)
  addrList.append(soup.select_one('.add1').string)

# telList
menuDF['전화번호'] = telList
menuDF['구주소'] = addrList
menuDF.drop(columns='주소',inplace=True)
menuDF

```

```python
# 업종이 한식인 가게의 정보만 출력
menuDF[menuDF['업종']=='한식']
```
