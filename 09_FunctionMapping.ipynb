# 함수 매핑 (mapping)

시리즈나 데이터프레임의 개별 요소를 특정 함수에 1:1로 대응시키는 과정을 의미

사용자가 직접 만든 함수 (lambda 함수 포함)를 적용시킬 수 있기 때문에 for 문이나 기본 함수로 처리하는 것보다 훨씬 효율적인 강점

```python
# seaborn에서 타이타닉 데이터를 불러와서 'age','fare'열만 인덱싱
#     => 'ten' 열을 만들어서 => 10을 넣기

titanic = sns.load_dataset('titanic')
df = titanic.loc[:,['age','fare']]
df['num'] = 10
df.head()
```

```python
# 값을 넣으면 10을 더해주는 함수
def add_10(n):
  return n+10
# 값을 2개 넣으면 그 둘을 곱해주는 함수 
def mul_two_obj(a,b):
  return a*b
# 값을 3개 넣으면 그 셋을 더해주는 함수
def add_three_obj(a,b,c):
  return a+b+c

print(add_10(15))
print(mul_two_obj(12,3))
print(add_three_obj(1,2,3))
```

```python
# 이 함수들을 apply() 함수를 이용해 시리즈 객체의 개별요소에 적용
sr1 = df['age'].apply(add_10) # age의 모든 요소 + 10
sr1.head()
```

```python
sr2 = df['fare'].apply(mul_two_obj,b=0.9)
sr2.head()
# 첫번째 파라미터 : 함수 
# 두번째 파라미터 : 값
```

```python
sr3 = df['fare'].apply(add_three_obj,b=10.1,c=20)
sr3.head()
```

```python
# 'num'열의 값들을 lambda 함수를 이용하여 add_10 함수 적용
# (lambda a,b: print(a*b))(10,20)
sr3 = df['num'].apply(lambda x: add_10(x))
sr3.head()
```

apply() vs applymap()

```python
df_map = df.applymap(add_10)  # 데이터 프레임의 모든 요소에 함수를 적용함
df_map.head()
```

# 시리즈 객체에 함수 매핑

DataFrame객체.apply(매핑함수, axis=값)

열 또는 행에 대해서 함수를 적용시킬 수 있음

```python
# 결측값을 확인하는 함수
def missing_value(series):
  return series.isnull()
```

```python
# 위의 함수를 데이터프레임의 열에 매핑(axis=1)
result = df.apply(missing_value,axis=1)
print(result)
print()
print(type(result))
result.describe()
```

```python
# 어떤 값을 넣었을 때 최대값에서 최소값을 뺀 값을 반환하는 함수
def min_max(x):
  return x.max() - x.min()
```

```python
# 이 함수를 데이터프레임에 열, 행에 대해서 적용해서 연산
# axis=0 (index 방향/ 행) 
#    생략 가능, 위아래(행들의 연산), 열에 대한 결과

# axis=1 (columns / 열 방향)
#   생략 불가능, 양옆(열들의 연산), 행에 대한 결과
result = df.apply(min_max, axis=0) #행
result1 = df.apply(min_max, axis=1) #열
print(df.head())
print()
print(result.head())
print()
print(result1.head())
```

```python
def add_two_obj2(a,b):
  return a+b
```

```python
# 람다 함수를 이용해서 위의 함수를 적용 -> 'age', 'num' 열의 데이터
# 새롭게 열 추가해서 결과 넣기
df['add'] = df.apply(lambda x: add_two_obj2(x['age'],x['num']), axis=1)
df.head()
```

# 3. 데이터 프레임 객체에 함수를 매핑

# DataFrame객체.pipe() 함수

pipe() 함수는 데이터프레임 객체에 바로 적용시키는 함수

파라미터에 들어가는 매핑함수의 반환 형태를 그대로 따라서 값을 반환


```python
# 타이타닉 데이터셋 불러와서 -> 'age', 'fare'열만 나오게
titanic = sns.load_dataset('titanic')
df = titanic.loc[:,['age','fare']]
df.head()
```

```python
# 값이 비어있는가에 대한 함수
def missing_value2(x):
  return x.isnull()
# 각 열의 비어있는 요소의 총 개수를 구하는 함수
def missing_count(x):
  return missing_value2(x).sum()
# 각 열의 비어있는 요소의 총 개수를 더한값을 구해주는 함수
def total_number_missing(x):
  return missing_count(x).sum()
```

```python
# 순차적 적용
result_df = df.pipe(missing_value2)
print(result_df)
print()
print(type(result_df))
```

```python
result_mc = df.pipe(missing_count)
print(result_mc)
print()
print(type(result_mc))
```

```python
result_total = df.pipe(total_number_missing)
print(result_total)
print()
print(type(result_total))
```
