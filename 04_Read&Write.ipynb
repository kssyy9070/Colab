# 1-3. 다양한 옵션

read_csv()의 다양한 옵션은 판다스 공식홈페이지에서 상세하게 보여줌

많이 쓰이는 옵션들 간단하게 정리(default 기준)

sep (',') : 자료의 구분 기준을 설정 

header ('infer') : 첫 행을 열이름으로 쓸 것ㅇ니지

names : 열 이름을 리스트로 입력해 줄 수 있음

index_col (None) : 특정 열 이름으로 인덱스를 지정

dtype : 개별 열 or 모든 열의 타입을 지정 

na_values : 결측값(NaN)으로 인식할 문자를 지정

keep_default_na (True) : 결측값을 포함할지 여부 결정

na_filter (True) : 결측값 탐지

skip_blank_lines (True) : 비어있는 줄은 결측값으로 판단하지 않고 건너뜀

nrow : 읽을 파일 행(row)의 수

여러개의 데이터프레임을 하나의 Excel 파일로 저장

Excel에는 csv 와는 다르게 sheet 라는게 존재

ExcelWriter을 활용하면 개별 sheet에 각각 데이터들 exprot

데이터분석에는 csv가 주로 사용되지만, 문서정리할 때 유용하게 사용되고 있음 


```python
data1 = pd.read_excel('ssec2203.xls',sheet_name='1.전체')
df2 = pd.DataFrame(data1)
df2
```

```python
data2 = pd.read_excel('ssec2203.xls',sheet_name='1.전체증감')
df3 = pd.DataFrame(data2)
df3
```

```python
# df2, df3을 엑셀 시트 1, 2로 각각 저장
writer = pd.ExcelWriter('test_excel1.xlsx')
df2.to_excel(writer, sheet_name = '1')
df3.to_excel(writer, sheet_name = '2')
writer.save()
```

pd.ExcelWriter('test_excel1.xlsx')를 해주게 되면

해당 파일에 writer 변수를 이용해서 쓸 준비가 됐다는 뜻 !

작업이 끝나면 writer.save()를 이용해 마무리를 하면

해당경로에 액셀파일이 저장

sheet_name='Sheet1' 옵션은 디폴드값이 Sheet1로 되어있기 때문에

따로 지정해주지 않으면 내용을 덮어쓰게 됨
