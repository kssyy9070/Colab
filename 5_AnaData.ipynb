# 3-5. 상관계수(문자열은 제외) : corr()

상관관계는 두 변수간의 상관성에 대한 정보로 -1 ~ 1 사이의 값
- 변수간 상관관계 : 변수간 흐름이 얼마나 비슷한가를 나타내주는 척도
- a가 증가하면 b도 증가하는가 / 감소하는가
- or 증가의 성향이 얼마나 비슷한가

를 알 수 있음

1에 가까울수록 두 지표가 비슷하게 증가 or 감소

-1에 가까울수록 하나 증가 / 하나 감소

0에 가까울수록 두 지표는 서로 관계가 없다

꼭 인과관계를 의미하지는 않음

이 지표가 독립변수와 종속변수로 만들어져야 하는데.....

데이터의 특징을 이해하는데 있어서 (EDA 과정 - Exploratory Data Analysis, 탐색적 자료 분석) 사용해 볼 만 하다.

corr()함수는 문자열을 제외한 변수간 matrix를 생성하고, 각 쌍들의 상관계수를 반환

DataFrame, Matrix 등 둘 다 2차원 형태의 데이터 테이블 형태

DataFrame은 각 요소가 다른 자료형이어도 되지만, Matrix는 각 요소의 자료형을 일치시켜야 하는 차이점이 있음

```python
i.corr()
```

```python
i[['bmi','charges']].corr()
```
