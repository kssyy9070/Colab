머신러닝 알고리즘 : data가 가진 특성(feature)들을 비교해서 data의 패턴을 찾아가는 것

# 정규화(Normalization)

* 데이터를 특정 구간으로 바꾸는 방법(ex : 0 ~ 1 or 0 ~ 100)
* 식 : (측정값 - 최소값) / (최대값 - 최소값)
* 데이터 군에서 특정 데이터가 가지는 위치를 알고 싶을 때 사용

# 표준화(Standardization)

* 데이터를 0을 중심으로 양쪽으로 데이터를 분포시키는 방법
* 식 : (측정값 - 평균) / 표준편차
* 이상치 처리하는데 효과적임

# SVM(Support Vector Machine)

고진적인 이진분류 머신러닝 기법 중 하나
* 이진분류 : 0 아니면 1

N차원의 공간을 (N-1)차원으로 나눌수 있도록 초평면을 찾는 분류 기법

상당히 높은 인식 성능 자랑

'가 or 나' 라는 2가지 패턴이 있을 때 이 패턴을 구분하는 방법을 찾는 것

=> 구분 할 수 있는 최적의 경계를 찾아내는 것 

마진 : 클래스끼리의 간격, 데이터 사이의 거리

SVM의 특징 : [마진최대화]

support vector(서포트 벡터) : 마진에서 가장 가까이에 위치한 각 클래스들의 데이터

장점 : 어떤 패턴을 가지고 있는지 모르는 경우에도 제대로 분류할 수 있는 확률이 높음

단점 : 계산량에 부담이 있음

# 선형 분리
* 마진 두 클래스를 분류할 수 있는 최대 마진의 경계를 찾아내는 것, 모든 훈련 데이터는 마진의 바깥쪽에 위치
* 단 하나의 오차도 허용되지 않음

# 이변수 데이터 (sns.pairplot)

* 파라미터로 전달되는 DataFrame의 열을 두개씩 짝을 지을 수 있는 모든 조합을 표현
  * 열은 정수/실수형이어야 함
  * 열이 3개라면 3 * 3 = 9개의 그리드를 만듦
* 각 그리드의 두 변수간의 관계를 보여주는 그래프를 하나씩 그려나가면서 같은 변수끼리 짝을 이루는 '대각선 방향'으로는 histogram을 그림
* 서로 다른 변수간에는 산점도

```python
import pandas as pd
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
titanic.tail()
```

```python
titanic2 = titanic[['survived','pclass','age','fare']]
titanic2.head()
```

```python
# pairplot
fig = sns.pairplot(titanic2)
# 대각선 방향으로는 하나의 열의 histogram을 표현
# 대각선 기준 위 아래는 서로 축이 변한 것 뿐이지 같은 내용
```

```python
# hue 옵션 추가 (카테고리별 비교)
fig = sns.pairplot(titanic2, hue='survived') # 생존자를 기준으로 pairplot을 그린 것
```

```python
sns.countplot(x=titanic['class'],hue=titanic['who'])
```

```python
# 마커 변경
sns.pairplot(titanic2, hue='survived', markers=['o','s']) #원형, 사각형 , 마름모 : D
```

```python
# columns 제한
sns.pairplot(titanic2, hue='survived', markers=['D','s'], vars=['pclass','age'])
```

```python
# 테마 색상
# palette
# husl, Set2, Set3, Paired, ...
sns.pairplot(titanic2,hue='survived',markers=['D','o'],palette='husl')
```

```python
# 높이 조절
sns.pairplot(titanic2, hue='survived', palette='Set3', height=3)
```

```python
# 대각선을 기준으로 한쪽만 그리기
sns.pairplot(titanic2, hue='survived', palette='cool', corner=True)
```

```python
# 투명도 설정
sns.pairplot(titanic2, hue='survived', palette='flare', height=3, plot_kws={'alpha':0.5})
```

```python
# 대각선을 기준으로 원하는 그래프 넣기
fig = sns.pairplot(titanic2, hue='survived', palette='husl',height=3, plot_kws={'alpha':0.3})

# 대각선 기준 위쪽으로 산점도 그리기
# regplot : scatterplot + lineplot을 합쳐놓은 그래프
# lineplot은 scatterplot의 경향을 예측하는 쪽으로 선이 그려짐
fig = fig.map_upper(sns.regplot)

# 대각선 기준 아래쪽으로
# kdeplot : 2차원 밀집도
fig = fig.map_lower(sns.kdeplot)

# 대각선 기준으로 histogram 그리기
fig = fig.map_diag(sns.distplot)
```
