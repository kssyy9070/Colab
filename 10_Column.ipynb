# 2. 열 분리 : Series객체.str.get(인덱스)

```python
flight = sns.load_dataset('flights')
flight.head()

# 영어를 숫자로 (month를 숫자로)
```

```python
month2int = {'Jan':1,
             'Feb':2,
             'Mar':3,
             'Apr':4,
             'May':5,
             'Jun':6,
             'Jul':7,
             'Aug':8,
             'Sep':9,
             'Oct':10,
             'Nov':11,
             'Dec':12
             }

flight['month'] = flight['month'].map(month2int)
flight.head()
```

```python
# day 컬럼 만들고 컬럼값을 모두 15로 설정
flight.insert(2,'day',15) #위치,이름,값
flight.head()
```

```python
# year, month, day (연월일) 합쳐서 날짜로
flight['date'] = pd.to_datetime(flight[['year','month','day']])
flight.head()
```

```python
# 붙였던 date를 뜯어서 year, month, day로 변경
flight['date'] = flight['date'].astype('str')
dates = flight['date'].str.split('-')
dates
```

```python
# dates변수는 리스트를 개별 요소로 갖는 시리즈 객체형태 
# str.get() 함수를 이용하면 각 행별로 인덱싱이 가능
flight['y'] = dates.str.get(0)
flight['m'] = dates.str.get(1)
flight['d'] = dates.str.get(2)
flight.head()
```

# 2-1. 매핑함수를 이용해서 열 분리

매핑함수와 for문을 활용해서 위와 같은 결과를 도출

```python
flight['date'] = flight['date'].astype('str') # string으로 형변환
col = ['y','m','d']
for i in range(len(col)):
  flight[col[i]] = flight['date'].apply(lambda x : x.split('-')[i])

flight.head()

# 좋은 방법은 아닌 듯 함
# series객체.str.get()가 더 낫다
```
